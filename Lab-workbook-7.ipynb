### 2. Read a sequence from a fasta file and translate the reverse complement of the same

```python
from Bio import SeqIO
sr=SeqIO.read("sequence.fasta","fasta")
s=sr.seq
rc=s.reverse_complement()
final=rc.translate()
print(final)
```

### 3. Input a formatted filename like fasta or genbank, load the contained sequence. Input a motif like TATA or GAATTC and display all the locations where this motif is located in the sequence.

```python
from Bio import SeqIO
fn=input("Enter file name without extention ")
fo=input("Enter file format ")
motif=input("Enter motif ")
fn=fn+"."+fo
sr=SeqIO.read(fn,fo)
s=sr.seq
i=0
pos=0
L=[]
while True:
    pos=s.find(motif,i)
    if pos ==-1:
        break
    L.append(pos)
    i=pos+len(motif)
print(L)
```
